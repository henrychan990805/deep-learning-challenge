## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'NAME'])
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = ['T25', 'T14', 'T29', 'T15', 'T17']

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"OTHER_APPLICATION_TYPE")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
application_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
application_df['CLASSIFICATION'].value_counts().loc[lambda x : x > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`

classifications_to_keep = ['C1000', 'C2000', 'C1200', 'C3000', 'C2100',"C7000", "C1700","C4000", "C5000","C1270", "C2700", "C2800"]
classifications_to_replace = application_df['CLASSIFICATION'].unique().tolist()
for x in classifications_to_keep:
  classifications_to_replace.remove(x)
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"OTHER_CLASSIFICATION")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df["AFFILIATION"] = application_df["AFFILIATION"].replace("Other", "Other_Affiliation")
application_df["USE_CASE"] = application_df["USE_CASE"].replace("Other", "Other_USE_CASE")
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
# Concert "AFFILIATION" to numeric
AFFILIATION = pd.get_dummies(application_df['AFFILIATION'])
CLASSIFICATION = pd.get_dummies(application_df['CLASSIFICATION'])
APPLICATION_TYPE = pd.get_dummies(application_df['APPLICATION_TYPE'])
ORGANIZATION = pd.get_dummies(application_df['ORGANIZATION'])
INCOME_AMT = pd.get_dummies(application_df['INCOME_AMT'])
USECASE = pd.get_dummies(application_df['USE_CASE'])
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = application_df.join(AFFILIATION).join(CLASSIFICATION).join(APPLICATION_TYPE).join(ORGANIZATION).join(INCOME_AMT).join(USECASE)
```

```python
application_df = application_df.drop(columns=['AFFILIATION', 'CLASSIFICATION', 'APPLICATION_TYPE', 'ORGANIZATION', 'INCOME_AMT', 'USE_CASE', "SPECIAL_CONSIDERATIONS"])
```

```python
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df.drop(columns=['IS_SUCCESSFUL']).values
Y = application_df['IS_SUCCESSFUL'].values
# Split the preprocessed data into a training and testing dataset
X_train,X_test, Y_train, Y_test = train_test_split(X,Y,random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

###Attempt 2

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
# Optimization: Increase number of nodes in each layer and Increased Epochs.
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 48
hidden_nodes_layer2 = 30

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,Y_train,epochs=200)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,Y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
